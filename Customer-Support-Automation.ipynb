```python
import warnings
warnings.filterwarnings('ignore')

import os
from crewai import Agent, Task, Crew
from crewai_tools import SerperDevTool, ScrapeWebsiteTool

# Set Mistral AI API key
mistral_api_key = "Your Api Key Here"
os.environ["MISTRAL_API_KEY"] = mistral_api_key
os.environ["OPENAI_API_BASE"] = "https://api.mistral.ai/v1"  
os.environ["OPENAI_MODEL_NAME"] = "mistral-medium" 

# Define Support Agent
support_agent = Agent(
    role="Senior Support Representative",
    goal="Be the most friendly and helpful support representative in your team",
    backstory=(
        "You work at crewAI and are providing support to {customer}, a super important customer. "
        "You must provide the best support, ensuring full and complete answers with no assumptions."
    ),
    allow_delegation=False,
    verbose=True
)

# Define Quality Assurance Agent
support_quality_assurance_agent = Agent(
    role="Support Quality Assurance Specialist",
    goal="Ensure the highest support quality for the team",
    backstory=(
        "You work at crewAI, reviewing support responses to ensure they are complete, accurate, "
        "and meet high-quality standards."
    ),
    verbose=True
)

# Define Tools
search_tool = SerperDevTool()
scrape_tool = ScrapeWebsiteTool()
docs_scrape_tool = ScrapeWebsiteTool(
    website_url="https://docs.crewai.com/how-to/Creating-a-Crew-and-kick-it-off/"
)

# Define Inquiry Resolution Task
inquiry_resolution = Task(
    description="{customer} reached out with an important question:\n{inquiry}\n\n"
                "{person} from {customer} needs help. Provide a complete and detailed response.",
    expected_output="A fully detailed response addressing all parts of the customer's inquiry, "
                    "including sources and external references if applicable.",
    tools=[docs_scrape_tool],
    agent=support_agent
)

# Define Quality Assurance Review Task
quality_assurance_review = Task(
    description="Review the response for {customer}'s inquiry to ensure it is comprehensive, "
                "accurate, and maintains a friendly tone.",
    expected_output="A polished, final response ready for the customer, ensuring completeness "
                    "and clarity.",
    agent=support_quality_assurance_agent
)

# Create the Crew
crew = Crew(
    agents=[support_agent, support_quality_assurance_agent],
    tasks=[inquiry_resolution, quality_assurance_review],
    verbose=2
)

# Define Inputs
inputs = {
    "customer": "HMA",
    "person": "Arham",
    "inquiry": "How do I set up API keys for different AI models?"
               
}

# Kickoff the Crew Process
result = crew.kickoff(inputs=inputs)

# Display the Result
from IPython.display import Markdown
Markdown(result)
```
